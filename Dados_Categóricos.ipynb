## üíº **Dados Categ√≥ricos - Vagas**

````python
import pandas as pd
import numpy

# Opa eai Leo, vou tentar fazer uma brincadeira aqui.
# eai
# Pensei em criar dados haver com vagas de emprego, j√° que o professor tocou muito nesse assunto recentemente.

# Essa parte aqui como voc√™ pode ver √© a parte b√°sica que j√° vimos.
dados = {
    'Vaga': ['Dev Python', 'Analista de Sistemas', 'Analista de Dados', 'Cientista de Dados', 'Dev Java', 'Dev Back-End'],
    'Nivel': ['Pleno', 'Senior', 'Pleno', 'Senior', 'Est√°gio', 'Junior'],
    'Modalidade': ['H√≠brido', 'Presencial', 'H√≠brido', 'Presencial', 'H√≠brido', 'H√≠brido']
}

# Criar a fun√ß√£o de DataFrame.
df = pd.DataFrame(dados)

# Pelo que vi, tem como transforma cada Dado central como categorias, vou ver se da certo (deu certo).
df['Modalidade'] = df['Modalidade'].astype('category') # Esse astype ele cria a categoria a parti da palavra semelhante
                                                                     # Assim salvando memoria RAM.

print("--- Tabela de Modalidade ---")
print(df['Modalidade'].value_counts().reset_index()) #Isso aqui vai apenas puxar a fun√ß√£o chamada Modalidade, e contar os valores presentes nela.
                                       # Ou seja vai contar quantos 'H√≠bridos' e 'Presenciais' est√£o presentes dentro dele.
                                       # Adicionar na linha do c√≥digo .reset_index() para uma saida mais limpa e organizada.

print("-----  Tabela -----")

#Literalmente printar a tabela inteira.
print(df)
````

### üìÑ **Nesse c√≥digo eu fiz o seguinte**

* Todo o momento eu nao usei codigos que usam Strings, porque fica mais leve o processo, isso √© uma informa√ß√£o extremamente util para um banco de dados mais robusto.

* O Resto foi destacar certas informa√ß√µes como por exemplo apenas Modalidades, mas todo o c√≥digo √© so print de dados especificos e geral. *(Ainda n√£o vi c√≥digo que adiciona depois, ou remove)*

---
```
df['Modalidade'] = df['Modalidade'].astype('category')
```
1. Esse c√≥digo puxa a fun√ß√£o chamada 'Modalidade' e diz, quero puxar apenas as
categorias, ou seja palavras com o mesmo nome, nisso ao inves de pegar 3 'H√≠bridos' e salvar na memoria, ele apenas amazena 1 'H√≠brido' e conta quantos tem com a mesma categoria.


# ‚ùì **O que isso ajuda?**
* Espa√ßo de mem√≥ria *(Performance)*
* Processamento mais r√°pido em Banco de Dados que possuem muitas linhas de c√≥digo.

---
```
print(df['Modalidade'].value_counts())
```
2. Esse c√≥digo falando de forma bem direta, ele conta quantos dados com a mesma nomeclatura possu√≠ dentro de um GRUPO de uma dataframe.
Ou seja no print vai sair: H√≠brido 3, Presencial 3.

Caso eu queira uma saida mais limpa eu adiciono:

```
.reset_index()
```



### üìà **Matplotlib baseado na tabela**

````python
import matplotlib.pyplot as plt

cont = df['Modalidade'].value_counts() # Puxo 'Modalidade' da tabela

plt.figure(figsize=(8, 5))
cont.plot(kind = 'bar', color = ['blue', 'green'])

plt.xlabel('Modalidade')
plt.ylabel('Quantidade')
plt.title('Quantidade de Vagas')
plt.xticks(rotation = 0)
plt.show()
# O motivo de eu querer criar um gr√°fico foi nessa l√≥gica que pensei:
# O codigo os dev's entendem, o gr√°fico qualquer um entende (al√©m de focar na parte visual)
````

## üìä **Salvar em CSV - Vagas**

````python
import pandas as pd

#Criar dados categoricos.

dados = {
    'Vaga': ['Dev Python', 'Analista de Sistemas', 'Analista de Dados', 'Cientista de Dados', 'Dev Java', 'Dev Back-End'],
    'Nivel': ['Pleno', 'Senior', 'Pleno', 'Senior', 'Est√°gio', 'Junior'],
    'Modalidade': ['H√≠brido', 'Presencial', 'Presencial', 'Presencial', 'H√≠brido', 'H√≠brido']
}

df = pd.DataFrame(dados)
df['Modalidade'] = df['Modalidade'].astype('category')

# Salvar como CSV
df.to_csv("dados.csv", index=False, encoding="utf-8") # O index serve para tirar a numera√ß√£o e deixar mais limpo a tabela.
                                                      # O encoding √© porque meu codigo est√° com caracteres especiais, ent√£o serve para ele entender.
print("Arquivo CSV criado com sucesso!")
````
