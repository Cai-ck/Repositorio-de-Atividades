### 3.3.**C√≥digo de mapa t√©rmico**

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Criar dados usando o Pandas
# Fonte desses dados s√£o do howlongtobeat
dados = {
    'Jogos': ["Fifa", "Borderlands 2", "Dmc 5", "Dark Souls"],
    'Horas': [571, 170, 115, 112],
}

# Criar dataframe e definir os dados de 'Jogo' como indice (para n√£o dar erro)
df = pd.DataFrame(dados).set_index('Jogos')

# C√≥digo que inverte as colunas, ou seja jogo vira coluna e metrica vira linha
df_visual = df.T

# Bobagens
ax = sns.heatmap(df_visual, annot = True, cmap = 'YlGnBu', fmt = "g",
                 linewidths = .5, cbar_kws = {'label': 'N√≠vel de Engajamento'})

plt.title('M√©tricas de desempenho por jogo', fontsize = 14)
plt.xlabel('')
plt.xticks(rotation = 0)
plt.yticks(rotation = 0)

plt.show()
````

#### 3.3.1.**O que aprendi**
* Novos c√≥digos de sns, obviamente n√£o sei a complexidade que seria criar um mapa t√©rmico em apenas matplot, mas achei f√°cil de entender, usar as tres bibliotecas foi interessante porque estou come√ßando a usar o pandas com mais frequencias, claro que n√£o usei a melhor m√©trica para representar esse gr√°fico, mas achei interessante de fazer.

### 3.4.**Gr√°fico de Barras - Dados Game Brasil**

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Dados baseados na Pesquisa Game Brasil 2024 e Digital Brazil 2024
# Link para achar esses dados:(https://gogamers.gg/o-gamer-no-brasil/pgb-2024/)
dados = {
    'Plataforma': ['Smartphone', 'PC / Desktop', 'Console', 'Tablet'],
    'Uso (%)': [78.9, 42.3, 39.3, 14.5]
}

df = pd.DataFrame(dados)

plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Aqui eu criei um c√≥digo diferente baseado no que eles pedem, pois vai modificar futuramente a forma de escrever
grafico = sns.barplot(x = 'Plataforma', y = 'Uso (%)', data = df, palette = 'viridis',
    hue = 'Plataforma',
    legend = False
)

# Aqui colocara os valores emcima das barras, isso √© para melhorar o visual somente
for p in grafico.patches:
    grafico.annotate(format(p.get_height(), '.1f') + '%',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9),
                     textcoords = 'offset points')

plt.title('Prefer√™ncia de Dispositivos de Jogos no Brasil (2024)', fontsize=14)
plt.ylim(0, 100) # Literalmente limite
plt.show()
````

####3.4.1**O que aprendi**

* O c√≥digo annotate, ele grava cada numero em um grupo ao vez de trazer um valor unico, al√©m disso esse formato inteiro cerve para trazer o seguinte, p.get_height(Pegar no gr√°fico a altura), + '%' adicionar porcentagem na numera√ß√£o final, pegar o x do  grafico e a sua largura e dividir por 2, ha e va √© servve para indentificar o centro, depois o tamanho do texto x e y.

```
grafico.annotate(format(p.get_height(), '.1f') + '%',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9),
                     textcoords = 'offset points')
```
* A √∫nica coisa que eu preciso aprender √© esse textcoords, acho eu que seria as cordenadas, pois quando voce n√£o adiciona essa linha de c√≥digo todas as n√∫mera√ß√µes ficam  presas em um √∫nico eixo.

```
plt.ylim(0, 100)
```
* Cria um limite no eixo y, assim n√£o podendo ultrapassar

### 3.5.**Gr√°fico Animado** *Extra Plotly*

````python
import seaborn
import plotly.express as px
import pandas as pd


dados = {
    'Plataforma': ['Smartphone', 'PC / Desktop', 'Console', 'Tablet'],
    'Uso (%)': [78.9, 42.3, 39.3, 14.5],
    'Cor': ['Mobile', 'PC', 'Console', 'Mobile']
}

df = pd.DataFrame(dados)

# Para codar o gr√°fico """"interativo"""" n√£o entendi como mexe direito XD
fig = px.bar(df,
             x='Plataforma',
             y='Uso (%)',
             color='Plataforma',
             title='Prefer√™ncia de Dispositivos no Brasil (Interativo - 2024)',
             text='Uso (%)',
             hover_data=['Uso (%)'],
             template='plotly_dark')

# Outra sintaxe por conta do plotly
fig.show()
````
