## 2Ô∏è‚É£ **Pr√°tica 02 - Tri√¢ngulo**

````python
# Pr√°tica Tri√¢ngulo
def classificar_triangulo():
    try:
        L1 = float(input("Digite o valor de L1: "))
        L2 = float(input("Digite o valor de L2: "))
        L3 = float(input("Digite o valor de L3: "))

        if L1 <= 0 or L2 <= 0 or L3 <= 0:
            print("Parab√©ns! Seu cpf foi cancelado por burrice extrema")
            return

        if (L1 + L2 > L3) and (L1 + L3 > L2) and (L2 + L3 > L1):

            if L1 == L2 == L3:
                print("Tri√¢ngulo Equil√°tero")
            elif L1 == L2 or L1 == L3 or L2 == L3:
                print("Tri√¢ngulo Is√≥sceles")
            else:
                print("Tri√¢ngulo Escaleno")

    except ValueError:
        print("Valor est√° errado tem que ser n√∫merooooooooooooooooo")

classificar_triangulo()
````

### **Soma Acumulada**

````python
# Soma acumulada

def soma_acumulada():
    soma = 0
try:
    num1 = int(input("Por favor, digite um n√∫mero: "))
    num2 = int(input("Por favor, digite o segundo n√∫mero: "))
    soma = num1 + num2
    print(f"O valor est√°: {soma}")

    while soma < 100:
      try:
          num1 = int(input("Por favor, digite novamente um n√∫mero: "))
          soma += num1
          print(f"O valor est√°: {soma}")

      except ValueError:
        print("Deu erro 2 o retorno")

except ValueError:
  print("Erro no inicio amigo kkkkk √© n√∫mero.")

soma_acumulada()
````

### üìà **IBGE Taxa de Desemprego [2024]** - *Segundo Gr√°fico*-*A Finalizar*



````python
import matplotlib.pyplot as plt

""" Mesmo a estrutura sendo feita por inputs
     os dados coletados do site: ('https://www.poder360.com.br/poder-economia/taxa-de-desemprego-vai-a-66-em-agosto/')
     Se quiser seguir os Dados corretamente: taxas = [8.0, 6.3, 6.6, 13.9, 11.1, 14.9, 7.9, 6.6]
             anos = [2012, 2013, 2014, 2017, 2019, 2021, 2022, 2024]
"""

taxas = []
anos = []
qtd = 8

print(f"Eu te obrigo a bota {qtd} ANOS e {qtd} TAXAS")

while len(anos) < qtd:
  try:
    decisao = input(f"{len(anos)+1}-Digite o ANO (ou Clique 's' para sair.): ")
    if decisao.upper().strip() == 'S':
       break

    anos_digi = int(decisao)
    taxas_digi = float(input(f"Digite a TAXA! correspondente a {anos_digi}: "))

    anos.append(anos_digi)
    taxas.append(taxas_digi)

  except ValueError:
    print("Erro, tente novamente colocando N√öMEROS v√°lidos.")

if anos:
 plt.figure(figsize=(9, 5))
 plt.plot(anos, taxas, marker = 'o', linestyle = '-', color = 'orange', label = 'Taxa de Desemprego')
 plt.xticks(anos, rotation = 45)
 plt.title("Rec√°uculo de Desemprego [2024]")
 plt.xlabel("Anos")
 plt.ylabel("Taxas")
 plt.legend()
 plt.grid(True)
 plt.tight_layout()
 plt.show()
````
